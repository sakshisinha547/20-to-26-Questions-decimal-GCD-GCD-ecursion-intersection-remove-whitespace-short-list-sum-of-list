```python
def intersection(list1:list ,list2:list) -> list:
    if not isinstance(list1,list) or not isinstance(list2,list):
        raise TypeError("input must be list")
    if len(list1) == 0 or len(list2) == 0:
        raise ValueError("input is empty")

    result = []

    for i in list1:
        if i in list2 and i not in result:
            result.append(i)
    return result

if __name__ == "__main__":
    list1 = [1,2,3,4,5,6,7]
    list2 = [3,45,67,45,6,7]
    try:
        print(intersection(list1,list2))
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
    

    
    
```

```python
def remove_whitespace(text : str):
    if not isinstance(text ,str):
        raise TypeError("Input must be string")
    prev_space = False
    result = ""
    for char in text.strip():
        if not char.isspace():
            result = result +char
            prev_space = False
        else:
            if prev_space == False:
                result = result + " "
                prev_space = True
    
    return result  

if __name__ == "__main__":
    text = "python   \t    is   \n     fun"
    try:
        print(remove_whitespace(text))
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
    
            
    
```

```python
def recursive(list1: list):
    if not isinstance(list1,list):
        raise TypeError("input must be list")
    for item in list1:
        if not isinstance(item,int):
            raise ValueError("all element must be an integer")
    if len(list1) == 0:
        return 0

    return list1[0]+ recursive(list1[1:])

if __name__ == "__main__":
    list1 = [ 23,34,65,7,8,2,56,78]
    try:
       print(recursive(list1))
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
    
        
```
